## Task 1 - Descriptive Statistics 

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

1. What is the mean English grade for all students in the dataset?

```python
df['english.grade'].mean()
```

2. What is the median math grade across all students?

```python
df['math.grade'].median()
```

3. What is the mode of the sciences grades, if it exists?

```python
df['sciences.grade'].mode()
```

4. What is the range of ages among the students?

```python
df['age'].max()-df['age'].min()
```

5. What is the average (mean) portfolio rating for all students?

```python
df['portfolio.rating'].mean()
```

6. What is the standard deviation of the math grades?

```python
df['math.grade'].std()
```

7. What is the variance of the English grades?

```python
df['english.grade'].var()
```

8. How many students have a sciences grade above the mean sciences grade?

```python
mean_science = df['sciences.grade'].mean()
students = (df['sciences.grade']>mean_science).sum()
print('Total student above the mean science grade = ',students)
```

9. What is the interquartile range (IQR) of the language grades?

```python
#IQR --> Q3-Q1

df['language.grade'].quantile(0.75)-df['language.grade'].quantile(0.25)
```

10. What is the midrange of the cover letter ratings?

```python
#midrange --> min+max/2

(df['coverletter.rating'].min()+df['coverletter.rating'].max())/2
```

11. Can you create a bar chart to compare the average grades (English, math, sciences, and language) across all students?

```python
avg_grades = df[["english.grade","math.grade","sciences.grade","language.grade"]].mean()
print(avg_grades)
```

```python
avg_grades = df[["english.grade","math.grade","sciences.grade","language.grade"]].mean()
avg_grades.plot(kind="bar")
plt.figure(figsize=(6,4))
avg_grades.plot(kind="bar")
plt.title("Average Grades Across Subjects")
plt.ylabel("Average Grade")
plt.xticks(rotation=0)
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.show()
```

12. Can you plot a box plot to show the spread and outliers in the cover letter ratings?

```python
plt.boxplot(df['coverletter.rating'])
plt.title('ccover Letter Rating Distribution')
plt.ylabel('Rating')
plt.show()
```
